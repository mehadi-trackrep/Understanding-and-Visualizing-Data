```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() #V.V.I. => for set(), we can use direct 'plt' variable instead of 'sns' explicitly!!
```

```python
url = 'nhanes_2015_2016-Copy1.csv'
df = pd.read_csv(url)
df.columns
```

```python
x = df.BMXWT.dropna().head(5660)
y = df.BMXHT.dropna().head(5660)
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,2)
# plt.hist(x=x, bins=15)
sns.distplot(x, bins=100)
plt.title('Marginal probability of x(BMXWT)')
 
plt.subplot(2,2,1)
# plt.hist(x=y, bins=15)
sns.distplot(y, bins=15)
plt.title('Marginal probability of y(BMXHT)')

plt.subplot(2,2,4)
# plt.hist(x=x, bins=15)
sns.distplot(x)
plt.title('Marginal probability of x1(BMXWT)')
 
plt.subplot(2,2,3)
# plt.hist(x=y, bins=15)
sns.distplot(y)
plt.title('Marginal probability of y1(BMXHT)')

plt.show()
```

```python
print(x, y)
```

```python
#V.V.I. => Joint Distribution... 
plt.figure(figsize=(16,10))
plt.subplot(2,2, 2)
plt.scatter(x=x, y=y)
plt.title('Joint Distribution of (BMXWT & BMXHT)')

plt.subplot(2,2,4)
plt.hist(x=x, bins=20)
plt.title('Marginal Distribution of x=BMXWT')

plt.subplot(2,2,1)
plt.hist(x=y, orientation='horizontal', bins=20)
plt.title('Marginal Distribution of y=BMXHT')
plt.show()
```

Let's find the association between (BMXWT & BPXSY1) & (BMXHT & BPXSY1)

```python
x1 = df.BMXWT.dropna().head(5401)
y1 = df.BPXSY1.dropna().head(5401)
# print(x1, y1)
```

```python
x2 = df.BMXHT.dropna().head(5401)
y2 = df.BPXSY1.dropna().head(5401)
# print(x2, y2)
```

```python
plt.figure(figsize=(20,15))

plt.subplot(4,4,2)
sns.scatterplot(x=x1, y=y1)
plt.title('Joint Distribtion of x1,y1')
plt.subplot(4,4,1)
plt.hist(x=y1, orientation='horizontal', bins=20)  #seaborn e orientation maybe nai!! :(
plt.title('Marginal Distribution of y1=BPXSY1')
plt.subplot(4,4,6)
# plt.hist(x=x1, bins=20) Or, (V.V.I.)
sns.distplot(x1)
plt.title('Marginal Distribution of x1=BMXWT')

plt.subplot(4,4, 4)
sns.scatterplot(x=x2, y=y2)
plt.title('Joint Distribution of x2,y2')
plt.subplot(4,4, 8)
# plt.hist(x=x2, bins=20)
sns.distplot(x2)
plt.title('Marginal Distribution of x2=BMXHT')
plt.subplot(4,4, 3)
plt.hist(x=y2, orientation='horizontal', bins=20)
plt.title('Marginal Distribution of y2=BPXSY1')
plt.show()
```
